```python
DATADIR = ""
DATAFILE = "C:/Users/VArun/Downloads/beatles-diskography.csv"

data = []
temp_array=[]
data_dict={}
def parse_file(datafile):
    with open(datafile, 'r') as f:
        f.next()
        for line in f:
            line=line.split(",")
            temp_array.append(line)

        for value in temp_array:
            data_dict['Title']=value[0]
            data_dict['Released']=value[1]
            data_dict['Label']=value[2]
            data_dict['UK Chart Position']=value[3]
            data_dict['US Chart Position']=value[4]
            data_dict['BPI Certification']=value[5]
            data_dict['RIAA Certification']=value[6]
            data.append(data_dict)

        return data

parse_file(DATAFILE)
```
